```python
import pandas as pd

# 중첩 리스트(배열)를 데이터프레임으로 변환하기

df = pd.DataFrame([['컴퓨터과학과', 1, '여'], ['국어교육과', 2, '여'], ['토목공학과', 3, '남'], ['간호학과', 4, '여'], ['물리학과', 4, '남']],
                  index = ['김지은', '박은하', '박길준', '이미도', '이진영'],
                  columns = ['학과', '학년', '성별'])

print('\n')
df
```

```python
df.rename(columns = {'학과' : '소속'}, inplace = True)
df.rename(index = {'김지은' : '김지영'}, inplace = True)
df
```

```python
df2 = df.copy()
df2.drop('이진영', inplace = True)
df2
```

```python
df3 = df.copy()
df3.drop(['성별'], axis = 1, inplace = True)
df3
```
