```python
bin_divider = [3.1, 10, 17, 24, 30]
bin_name = ["쌀쌀해", "적당해", "딱좋아", "더울걸"]

bicycle['cut_temperature'] = pd.cut(bicycle['hour_bef_temperature'],
                                    bins = bin_divider,
                                    labels = bin_name,
                                    include_lowest = True)
bicycle[['hour_bef_temperature', 'cut_temperature']]
```

```python
bicycle['cut_temperature'].value_counts()
```

```python
bicycle['qcut_temperature'] = pd.qcut(bicycle['hour_bef_temperature'], 4, labels = ["쌀쌀해", "적당해", "딱좋아", "더울걸"])
bicycle['qcut_temperature'].value_counts()
```
