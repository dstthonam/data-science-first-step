```python
import pandas as pd

# 딕셔너리를 시리즈로 변환하기
pl ={'C' : 5, 'JAVA' : 8, 'JS' : 7, 'PYTHON' : 20}

sr = pd.Series(pl)

print(type(sr))
print(sr)
print()

#리스트를 시리즈로 변환하기
pl_list = ['C', 'JAVA', 'JS', 'PYTHON']

sr = pd.Series(pl_list)

print(type(sr))
print(sr)
```

```python
import pandas as pd

myinfo_dic ={'이름' : '박인성',
             '학과' : '사회학과',
             '학번' : '2018010436',
             '성별' : '남성'}

sr = pd.Series(myinfo_dic)

print(sr)
print()
print("이름: ", sr['이름'])
print()
print(sr[['이름', '성별']])
print()

myinfo_list = ['박인성', '사회학과', '2018010436', '남']

sr = pd.Series(myinfo_list)

print(sr)
print()
print("이름: ",sr[0])
print()
print(sr[[0,3]])
#print(sr[])
```
