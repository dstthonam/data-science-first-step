# 타이타닉 데이터로 해보기


```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

#결측 데이터 확인 
df.info()   #  age, embarked , deck, embark_town 에 결측치 존재

```

```python
print(df.isnull().sum()) #  age, embarked , deck, embark_town 에 결측치 존재
```

```python
# 결측 데이터 제거  - deck는 889개의 데이터 중  688개의 데이터가 누락 ==> 너무나 많은 데이터가 누락되었으므로 열 삭제하기로 하자.
df = df.dropna(axis=1, thresh=500)
print(df.columns, '\n')

print(df.isnull().sum()) 
```

```python
# 결측 데이터 변경 1  - age : 177  ==> 평균으로 변경하기

age_mean = df['age'].mean(axis=0) 

df['age'].fillna(age_mean, inplace=True)

print(df.isnull().sum()) 
```

```python
# 결측 데이터 변경 2 - embarked : 2  ==> 최빈값으로 변경하기

# 최빈값 확인하기
print(df['embarked'].value_counts())
```

```python
df['embarked'].fillna("S", inplace=True)

print(df.isnull().sum()) 
```

```python
max_fre = df['embark_town'].value_counts(dropna=True).idxmax()
max_fre
```

```python
# 결측 데이터 변경 2 - embark_town : 2  ==> 최빈값으로 변경하기

max_fre = df['embark_town'].value_counts(dropna=True).idxmax()
df['embark_town'].fillna(max_fre, inplace=True)

print(df.isnull().sum()) 
```

# 범주형 데이터로 변환하기 - 따릉이 데이터 셋

/content/drive/MyDrive/data/bicycle.csv

```python
import pandas as pd
 

bicycle = pd.read_csv('/content/drive/MyDrive/data/bicycle.csv')
```

```python
bicycle.info()
```

```python

bicycle.isnull().sum()   #1459개의 데이터 중 누락값이 117이 최대, 
```

```python
bicycle['hour_bef_temperature'].describe()  # 동일 구간으로 나누기 위해
```

```python
#cut() 메소드 이용하여 연속형 수치형 데이터를 특정 구간으로 나누기
#pd.cut( 데이터, 구간개수, 레이블명 )


bin_divider = [ 3.1, 12.8, 16.6, 20,  30]
bin_name = ['쌀쌀해','적당해', '딱좋아', '더울걸']

bicycle['cut_temperature'] = pd.cut( bicycle['hour_bef_temperature'], bins = bin_divider , labels=bin_name , include_lowest = True)

bicycle[['hour_bef_temperature', 'cut_temperature']]
```

```python
bicycle['cut_temperature'].value_counts()
```

```python
# 이건 제외할 것임
#  qcut() 메소드 이용하여 수치형 데이터를 특정 구간으로 나누기
#    : cut 함수와 다르게 동일한 길이로 구간을 나누는 개념이 아니라 동일한 개수로 구간을 나누는 함수이다.
#pd.qcut( 데이터, 구간개수, 레이블명 )

bin_name = ['쌀쌀해', '딱좋아', '더울걸']

bicycle['bin_temperature_q'] = pd.cut( bicycle['hour_bef_temperature'], 3 , labels=bin_name )

bicycle['bin_temperature_q'].value_counts()

```

# 시계열 데이터 처리하기 - 지진데이터 사용

```python
import pandas as pd
 
earthquake= pd.read_csv('/content/drive/MyDrive/data/earthquake.csv')
```

```python
earthquake.head()
```

```python
earthquake.info()
```

```python
#다른 자료형을 시계열 객체로 변환하기
earthquake["발생시각"] = pd.to_datetime(earthquake["발생시각"])
earthquake.info()
```

```python
#다른 자료형을 시계열 객체로 변환하는 또 다른 방법
earthquake["발생시각"]  = earthquake["발생시각"].astype("datetime64") #또는
```

```python
# 시계열 데이터의 날짜를 연, 월, 일, 시, 분으로 분리하기

earthquake["year"] = earthquake["발생시각"].dt.year
earthquake["month"] = earthquake["발생시각"].dt.month
earthquake["day"] = earthquake["발생시각"].dt.day
earthquake["hour"] = earthquake["발생시각"].dt.hour
earthquake["minute"] = earthquake["발생시각"].dt.minute

earthquake.head()
```

```python
# 연도별 최대 규모를 구하면
earthquake.groupby("year")["규모"].max()
```
