agg() 이용하여 그룹에 함수 적용
- agg(사용자정의 함수명)
- agg('집계함수명')
- agg([ 여러개 함수명 나열 ] ) : 모든열에 여러개 함수 적용
- agg({ 열1:함수1, 열2:함수2 ... } ): 각각 열에 각자 다른 함수 적용

```python
def min_max(x) :
    return x.max() - x.min()

groupedf = gapminder.groupby("continent")
agg_groupedf = groupedf.agg(min_max)
print(agg_groupedf)
```

```python
groupedf = gapminder.groupby("continent")[['lifeExp', 'pop', 'gdpPercap']]
agg_groupedf = groupedf.agg('mean')
print(agg_groupedf)
```

```python
groupedf = gapminder.groupby("continent")
agg_groupedf = groupedf.agg(['mean', 'max', 'min'])
print(agg_groupedf)
```

```python
groupedf = gapminder.groupby("year")[['lifeExp', 'pop', 'gdpPercap']]
agg_groupedf = groupedf.agg({'lifeExp' : 'max', 'pop' : 'min', 'gdpPercap' : 'mean'})
print(agg_groupedf)
```

# 데이터 결합하기


```python
import pandas as pd 
 
df1 = pd.read_csv('/content/drive/MyDrive/data/chart01.csv', sep='\t') 
df2 = pd.read_csv('/content/drive/MyDrive/data/chart02.csv', sep='\t') 
df3 = pd.read_csv('/content/drive/MyDrive/data/chart03.csv', sep='\t') 
df4 = pd.read_csv('/content/drive/MyDrive/data/chart04.csv', sep='\t') 

print(df1)
print(df2)
print(df3)
print(df4)
```

```python
# concat ()

concat_df1 = pd.concat([df1, df2])
concat_df1
```

```python
concat_df1 = pd.concat([df1, df2], ignore_index=True)
concat_df1
```

```python
concat_df2 = pd.concat([df1, df3], axis=1)
concat_df2
```

```python
# merge() - 병합, 기준이 되는 키가 양쪽 데이터프레임에 존재해야 됨
merge_df1 = pd.merge(df1, df3, how='outer', on='매장')
merge_df1
```

```python
merge_df2 = pd.merge(df1, df3, how='inner', on='매장')
merge_df2
```

```python
merge_df3 = pd.merge(df3, df4, how='left', on='매장')
merge_df3
```

```python
merge_df4= pd.merge(df4, df3, how='left', on='매장')
merge_df4
```

```python
merge_df3 = pd.merge(df3, df4, how='right', on='매장')
merge_df3
```
