```python
#CSV 파일로 저장하기
print(df1)
print('\n')
print(df2)
print('\n')

df_to_concat = pd.concat([df1, df2], ignore_index=True)
print(df_to_concat)

df_to_concat.to_csv("df_to_csv.csv")
```

```python
#excel 파일로 저장하기
print(df1)
print('\n')
print(df2)
print('\n')

df_to_concat = pd.concat([df1, df2], ignore_index=True)  #하나의 파일로 합쳐서 저장
df_to_concat.to_excel("df_to_excel.xlsx")

writer = pd.ExcelWriter("df_to_excelwriter.xlsx")
df1.to_excel(writer, sheet_name="sheet1")
df2.to_excel(writer, sheet_name="sheet2")
writer.save()
```
