```python
# 데이터 선택(추출)           DataFrame 객체.loc[행 인덱스, 열 인덱스] or DataFrame 객체.iloc[행 번호, 열 번호]
# 1행과 2개이상 열 선택 , 2개이상 행과 1열을 선택하면 시리즈로 반환
# 2개이상 행과 2개 이상 열로 선택하면 데이터프레임으로 반환



df = pd. DataFrame([['컴퓨터과학과',1, '여'],
                    ['국어교육과', 2,  '여'],
                    ['토목공학과',  3,  '남'], 
                    ['간호학과', 4, '여'], 
                    ['물리학과', 4, '남']],
                    index = ['김지은', '박은화', '박길준', '이미도','이진영'],  
                    columns = ['학과', '학년', '성별'])

major3 = df.loc['박은화', '학과']   # 한개 행 선택
major4 = df.iloc[1, 0]
print(major3, major4)
print('\n')

major5 = df.loc['박은화', ['학과','학년' ]] #1 행과 2 열 선택
major6 = df.iloc[1, [0, 1]]
print(major5)
print('\n')
print(major6)
print('\n')

major7 = df.loc[['박은화', '이미도'], ['학과','학년' ]] #2 행과 2 열 선택
major8 = df.iloc[[1,3],  [0, 1]]
print(major7)
print('\n')
print(major8)
```

 인덱스 활용 -  새로운 인덱스로 재지정하기.

 reindex() 사용.

     DataFrame 객체.reindex(새로운 인덱스)

- 기존 데이터프레임을 변경하지 않고 새로운 데이터프레임 객체를 생성
기존 데이터프레임에 존재하지 않는 행 인덱스는 그대로 두고
기존 데이터프레임에 존재하지 않는 행 인덱스가 새롭게 추가되는 경우 그 행의 데이터 값은 NaN 값이 입력된다. (NaN : Not a Number)
- fill_value 옵션을 사용하면 원하는 값을 입력할 수 있음.



```python
import pandas as pd
```

```python
dict_data = {'이름':['김지은', '박은화', '박길준', '이미도','이진영'],
            '학과': ['컴퓨터과학과', '국어교육과', '토목공학과', '간호학과', '물리학과'],	
            '학년':	[1, 2, 3, 4, 4],
            '성별': ['여','여','남','여','남']}

df = pd. DataFrame(dict_data )
print(df)
print('\n') 

df = df.set_index('이름')
print(df )
print('\n')

new_index = ['김지은', '박은화', '박길준', '이미도','이진영', '신진수']
new_df2 = df.reindex(new_index)
print(new_df2)
print('\n')

new_df3 = df.reindex(new_index, fill_value='0')
print(new_df3)
print('\n')
```

```python
# 데이터프레임 연산
# 두가지 방법 - 데이터프레임과 숫자를 연산하는 것. 데이터프레임과 데이터프레임의 데이터를 연산하는 것.

# 2. 데이터프레임과 데이터프레임의 데이터를 연산하는 것.
# 같은 행, 열에 위치한 데이터끼리 연산을 한다. 새로운 데이터프레임을 만들어 동일한 위치에 결과를 저장. 
# 어느 한 쪽에 데이터가 비어 있거나 NaN이면 연산 결과는 NaN으로 처리한다.

# DataFrame 객체 +  DataFrame 객체

df1 = pd. DataFrame([[84, 87, 90], [89, 58, 67], [98, 93, 80], [89, 79, 94], [98, 80, 92]],
                    index = ['stu01', 'stu02', 'stu03', 'stu04','stu05'],  
                    columns = ['1차', '2차', '3차'])

df2 = pd. DataFrame([[77, 97], [86, 68], [78, 83], [79, 69], [85, 93]],
                    index = ['stu01', 'stu02', 'stu03', 'stu04','stu05'],  
                    columns = ['1차', '2차'])


print(df1)
print('\n')
print(df2)
print('\n')

new_df = (df1 + df2) /2

print(new_df)
```
