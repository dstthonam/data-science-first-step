# 통계함수 적용하기

```python
# 통계함수 적용하기 - max 함수


#숫자값 비교하여 큰 값
#문자열은 아스키코드 값으로 변환하여 크고 작음을 비교
print(gapminder.max())   # 모든 열에 대해서 가장 큰 값을 찾아서 반환
print('\n')
print(gapminder['lifeExp'].max())
print('\n')
print(gapminder['country'].max())
 
```

```python
# 통계함수 적용하기 - min 함수

print(gapminder.min())
print('\n')
print(gapminder['lifeExp'].min())
print('\n')
print(gapminder['country'].min())
```

```python
# 통계함수 적용하기 - mean 함수  평균값  
# 수치데이터를 가지고 있는 열에 대해서 계산하여 시리즈 객체로 반환

print(gapminder.mean())
print('\n')
print(gapminder['lifeExp'].mean())
```

```python
 # 통계함수 적용하기 - median 함수  중간값
 #수치데이터를 가지고 있는 열에 대해서 계산하여 시리즈 객체로 반환

print(gapminder.median())
print('\n')
print(gapminder['lifeExp'].median())
```

```python
# 통계함수 적용하기 - std 함수   표준편차

print(gapminder.std())
print('\n')
print(gapminder['lifeExp'].std())
```

```python
# 통계함수 적용하기 - corr 함수   상관계수
# 2개의 열에 대한 상관계수를 계산

print(gapminder.corr())
print('\n')
print(gapminder[['lifeExp' , 'gdpPercap']].corr())
```

문제 1)pop 평균보다 인구가 높은 국가의 2000년대 데이터를 필터링하고,continent, country, pop, gdpPercap 열만 출력하라

```python
gapminder['year'].unique() #우선 year데이터에 어떤 값이 있는지 확인합니다. 
print(gapminder['year'].unique())
```

```python
#2000년대 데이터는 2002년과 2007년이 있네요.

gapminder1=gapminder[(gapminder['pop']>=gapminder['pop'].mean())&(gapminder['year'].isin([2002, 2007]))]
print(gapminder1)
```

```python
gapminder1[['continent','country', 'pop', 'gdpPercap']]
```

문제 2 ) 문제1에서 추출한 데이터프레임에서 한국, 중국, 일본의 데이터를 추출하고, 국가별로 그룹화해서 인구수의 평균을 구하라.

```python
gapminder2=gapminder1[['continent','country', 'pop', 'gdpPercap']]
print(gapminder2)
```

```python
gapminder2.country.unique()
```

```python
#값을 확인해서 east_asia에 넣어줍니다.
east_asia=['Korea, Rep.','Japan','China']
```

```python
gapminder2[gapminder2['country'].isin(east_asia)]
```

```python
gapminder2[gapminder2['country'].isin(east_asia)].groupby(['country']).mean()['pop']
```

문제3. 문제2에서 작성한 데이터프레임으로 3개국의 2000년대 인구수를 막대그래프로 작성하라(plot()을 사용할 것)

```python
gapminder3=gapminder2[gapminder2['country'].isin(east_asia)].groupby(['country']).mean()['pop']

gapminder3.plot(kind='bar',title="2000's population in East_asia", rot=1, color=['red', 'blue', 'green'])
```
