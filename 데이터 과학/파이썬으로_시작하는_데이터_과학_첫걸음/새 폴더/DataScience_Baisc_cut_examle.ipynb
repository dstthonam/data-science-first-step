```python
import pandas as pd

 
df = pd.DataFrame({"점수": [49,50,53,54,60,70,80,90,97,89,95,96,100,95,73,95,86,93,85,95]})
df
```

```python

bin_divider = [ 50, 60, 70, 80, 90, 100]
bin_name = ['F', 'D','C', 'B', 'A']

df['학점'] = pd.cut(df['점수'] ,bins = bin_divider ,labels=bin_name,  include_lowest = True )
df
```
