**seaborn**

Matplotlib의 기능과 스타일을 확장한 파이썬 시각화 도구의 고급 버전으로
비교적 단순한 인터페이스를 제공하기 때문에 초보자도 어렵지 않게 사용할 수 있음.

데이터프레임으로 다양한 통계 지표를 낼 수 있는 시각화 차트를 제공하기 
때문에 데이터 분석에 활발히 사용되고 있는 라이브러리


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tips_df = sns.load_dataset('tips')
tips_df
```

```python
# histplot() 메소드를 이용하여 범위에 포함하는 관측수를 세어 그래프에 나타낼 수 있다.
sns.histplot(data=tips_df, x='total_bill')
#plt.show()
```

```python
## countplot() 메소드를 이용하여 각 범주에 속하는 데이터의 개수를 막대 그래프로 나타낼 수 있다

sns.countplot(data=tips_df, x='sex')
#plt.show()
```

```python
sns.countplot(data=tips_df, x='day')
#plt.show()
```

```python
# barplot() 메소드를 이용하여 각 범주에 속하는 데이터의 평균값을 막대그래프로 나타낼 수 있다.

sns.barplot(data=tips_df, x='day', y='total_bill', palette='Set1')
#plt.show()
```

```python
sns.barplot(data=tips_df, x='day', y='tip',hue='sex',  palette='Set1')   # ci=False 막대위 검은선 없앰
#plt.show()
```

```python
# boxplot() 메소드를 이용하여 최대, 최소, 평균, 1사분위수, 중간값, 3사분위수, 그리고 특이치(outliar) 등 다양한 통계량을 나타낼 수 있다.
sns.boxplot(data=tips_df, x='smoker', y='total_bill',hue='sex',  palette='Set2')
#plt.show()
```

```python
# violinplot() 메소드를 이용하여 데이터의 비교 분포도를 나타낼 수 있다.

sns.violinplot(data=tips_df,x='smoker', y='total_bill')
#plt.show()
```

```python
# scatterplot() 메소드를 이용하여 연속형 변수 사이의 관계 분포도를 나타낼 수 있다.

sns.scatterplot(data=tips_df,x= 'total_bill', y='tip')
#plt.show()
```

```python
sns.scatterplot(data=tips_df,x= 'total_bill', y='tip', hue='sex')
#plt.show()
```

```python
sns.scatterplot(data=tips_df,x= 'total_bill', y='tip', hue='day')
#plt.show()
```

```python
#  catplot() 메소드를 이용하여 범주형 변수의 분포도를 나타낼 수 있다.

sns.catplot(data=tips_df,x='day', y='total_bill' )
#plt.show()
```

```python
#  catplot() 메소드를 이용하여 범주형 변수의 분포도를 나타낼 수 있다.

sns.catplot(data=tips_df,x='day', y='total_bill', kind='swarm' )
#plt.show()
```

```python
# heatmap() 메소드를 이용하여 상관관계의 결과를 행렬 형태로 나타낼 수 있다

sns.heatmap(data=tips_df.corr(), annot =True )
#plt.show()
```

```python
# FacetGrid() 메소드를 이용하여 다양한 범주형 값을 가지는 데이터를 행, 열 방향으로 서로 다른 조건을 적용하여 여러 개의 서브 그래프를 제작할 수 있다.

facet = sns.FacetGrid(tips_df, col='time', row='smoker', hue='sex')
facet.map(sns.scatterplot, 'total_bill', 'tip')
facet.add_legend()
#plt.show()
```
