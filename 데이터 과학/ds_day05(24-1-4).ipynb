```python
gapminder = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/데이터 과학/파이썬으로_시작하는_데이터_과학_첫걸음/gapminder.csv')
gapminder.info()
print('\n')
gapminder['year'].unique()

pop_mean = gapminder["pop"].mean()
print(pop_mean)
print('\n')

gapminder_1 = gapminder[(gapminder["pop"] >= pop_mean)]
gapminder_2 = gapminder_1[gapminder_1['year'].isin([2002, 2007])]
groupA = gapminder_2[['continent', 'country', 'pop', 'gdpPercap']]
print(groupA)
```

```python
groupA.country.unique()
```

```python
groupA_data = groupA[groupA.country.isin(['Japan', 'Korea, Rep.', 'China'])]
groupA_data_mean = groupA_data.groupby('country')["pop"].mean()
groupA_data_mean.plot(kind = 'bar', title = "2000's population in East_asia", rot = 1, color = ['red', 'blue', 'green'])
```

```python
bicycle = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/데이터 과학/파이썬으로_시작하는_데이터_과학_첫걸음/bicycle.csv')
bicycle.columns = ['id', 'hour', 'temperature', 'precipitation', 'windspeed', 'humidity', 'visibility', 'ozone', 'pm10', 'pm2.5', 'count']
print(bicycle.head())
```

```python
bicycle.info()
```

```python
bicycle.isnull().sum()
```

```python
bicycle = bicycle.dropna(axis = 0)
print(bicycle.isnull().sum())
```

```python
bicycle.corr(method = 'pearson')
# bicycle.corr() 도 가능
```

```python
bicycle.plot(x = 'temperature', y = 'count', kind = 'scatter')
```

```python
bicycle.plot(x = 'precipitation', y = 'count', kind = 'scatter')
```

```python
import pandas as pd
movie_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/데이터 과학/파이썬으로_시작하는_데이터_과학_첫걸음/movies_train.csv')
movie_df.head()
```

```python
movie_df.info()
```

```python
ten_million_df = movie_df[movie_df["box_off_num"] >= 10000000][['title', 'box_off_num']]
ten_million_df
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'NanumGothic')
plt.bar(ten_million_df ['title'],
        ten_million_df ["box_off_num"],
        color = 'blue')
```

```python
plt.figure(figsize = (8, 6))

plt.rc('font', family='NanumGothic')
plt.bar(ten_million_df['title'], ten_million_df["box_off_num"], color = 'blue')

plt.title("천만 관중을 달성한 영화", fontsize = 16)
plt.xlabel("영화제목", fontsize = 12)
plt.ylabel("관객수", fontsize = 12)
plt.grid(True)
```
